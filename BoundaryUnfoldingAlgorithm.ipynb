```python
# Boundary unfolding transformation approach
import matplotlib.pyplot as plt
import cv2
import math
import numpy as np
import statsmodels.api as sm
from statsmodels.graphics import tsaplots
from statistics import mean
#an example on calcuating boundary unfolding graph
img = cv2.imread('../input/image.jpg',0)  #the image is the convered binary image 
img= cv2.resize(img, (200,200), interpolation = cv2.INTER_AREA)
img_bw=img<=128
img_bw=255*img_bw.astype('uint8')
# this part is to determine the contour of the shape
contours, _ = cv2.findContours(img_bw,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_NONE)
ref = np.zeros_like(img)
cv2.drawContours(ref, contours, 0, 255, 1)
# find center of gravity using moment
M = cv2.moments(contours[0])
cx = int(M['m10']/M['m00'])
cy = int(M['m01']/M['m00'])
print(cx)
print(cy)
# Get dimensions of the image
width = img.shape[1]
height = img.shape[0]
N = 360     # the total optimum number of angles (or boundary points) we want 
dis=np.zeros((N,1))
for i in range(N):
  tmp = np.zeros_like(img)
  theta = i*(360/N)
  theta =math.radians(theta)
  cv2.line(tmp, (cx, cy),                     #line from centroid to each boundary point to calculate distance
           (int(cx+math.cos(theta)*width),
            int(cy-math.sin(theta)*height)), 255,2)
  (row,col) = np.nonzero(np.logical_and(tmp, ref))  #x and y coordinates of boundary points
  col1=mean(col)     
  x=(col1-cx)    
  row1=mean(row)     
  y=(row1-cy)
#compute unfolded signal/graph
  dis[i]=math.sqrt(((x)**2)+((y)**2))
print(dis)
plt.figure('unfolded signal')
plt.plot(dis)
plt.xlabel('theta')
plt.ylabel('r (theta)')
#compute autocorrelation function
autocorr=sm.tsa.acf(dis, nlags=359)
plt.figure('Autocorrelation')
plt.plot(autocorr)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
```
