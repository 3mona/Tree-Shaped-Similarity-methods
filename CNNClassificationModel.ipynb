```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="whitegrid")
import os
import glob as gb
import cv2
import tensorflow as tf
import keras
#code given to each cluster
code = {'Columnar':0 ,'Irregular':1,'Open':2,'Oval':3,'Pyramidal':4,'Round':5,'Spreading':6}
def getcode(n) : 
    for x , y in code.items() : 
        if n == y : 
            return x    
X_train = []
y_train = []
for folder in  os.listdir('../input/tree-shaped-networks/Tree-Shaped Networks/TrainTrees') : 
    files = gb.glob(pathname= str('../input/tree-shaped-networks/Tree-Shaped Networks/TrainTrees//' + folder + '/*.jpg'))
    for file in files: 
        image = cv2.imread(file)
        image_array = cv2.resize(image , (100,100))
        X_train.append(list(image_array))
        y_train.append(code[folder])
plt.figure(figsize=(20,20))
#plot training images with cluster name
for n , i in enumerate(list(np.random.randint(0,len(X_train),49))) : 
    plt.subplot(7,7,n+1)
    plt.imshow(X_train[i])   
    plt.axis('off')
    plt.title(getcode(y_train[i]))
X_test= []
files = gb.glob(pathname= str('../input/tree-shaped-networks/Tree-Shaped Networks/TestTrees/*.jpg'))
for file in files: 
    image = cv2.imread(file)
    image_array = cv2.resize(image , (100,100))
    X_test.append(list(image_array))   
X_train = np.array(X_train)
X_test = np.array(X_test)
y_train = np.array(y_train)
#CNN Model
KerasModel = keras.models.Sequential([
        keras.layers.Conv2D(100,kernel_size=(3,3),activation='relu',input_shape=(100,100,3)),
        keras.layers.Conv2D(50,kernel_size=(3,3),activation='relu'),
        keras.layers.MaxPool2D(4,4),
        keras.layers.Conv2D(120,kernel_size=(3,3),activation='relu'),    
        keras.layers.Conv2D(80,kernel_size=(3,3),activation='relu'),    
        keras.layers.Conv2D(50,kernel_size=(3,3),activation='relu'),
        keras.layers.MaxPool2D(4,4),
        keras.layers.Flatten() ,    
        keras.layers.Dense(120,activation='relu') ,    
        keras.layers.Dense(100,activation='relu') ,    
        keras.layers.Dense(50,activation='relu') ,        
        keras.layers.Dropout(rate=0.5) ,            
        keras.layers.Dense(7,activation='softmax') ,    
        ])
KerasModel.compile(optimizer ='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
ThisModel = KerasModel.fit(X_train, y_train, epochs=40,batch_size=64,verbose=1)
y_result = KerasModel.predict(X_test)
plt.figure(figsize=(12,12))
#plot test images with cluster name
for n , i in enumerate(list(np.random.randint(0,len(X_test),120))) : 
    plt.subplot(12,10,n+1)
    plt.imshow(X_test[i])    
    plt.axis('off')
    plt.title(getcode(np.argmax(y_result[i])))
```
